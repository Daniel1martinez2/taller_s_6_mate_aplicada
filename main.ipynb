## 2.

(a) $a_n = \frac{1}{3^n}, n = 0,1,2,\dots$  
    <span style="color:red">----</span>Converge a $0$, ya que $\lim_{n \to \infty} \frac{1}{3^n} = 0$.

(b) $a_n = \pi^n, n = 0,1,2,\dots$  
    <span style="color:red">----</span>Diverge, ya que $\pi^n$ crece indefinidamente conforme $n \to \infty$.

(c) $b_n = \frac{3n-1}{2n+6}, n = 1,2,3,\dots$  
    <span style="color:red">----</span>Converge a $1$. $\lim_{n \to \infty} \frac{3n-1}{2n+6} = 1$.

(d) $c_n = \cos(\pi n), n = 0,1,2,\dots$  
    <span style="color:red">----</span>Diverge, ya que alterna entre $1$ y $-1$ conforme $n$ cambia.

(e) $a_n = \frac{3n^4 - 6n^3 + 1}{5n^3 + n^2 + 2}, n = 1,2,3,\dots$  
    <span style="color:red">----</span>Diverge, ya que $\lim_{n \to \infty} a_n = \infty$ porque el término dominante es $n^4$.

(f) $b_n = \frac{2n^5 - 3n^2 + 23}{7n^5 + 13n^4 - 12}, n = 1,2,3,\dots$  
    <span style="color:red">----</span>Converge a $\frac{2}{7}$. $\lim_{n \to \infty} \frac{2n^5 - 3n^2 + 23}{7n^5 + 13n^4 - 12} = \frac{2}{7}$.

(g) $c_n = \frac{45 - 16n^2}{13 + 5n + 6n^3}, n = 1,2,3,\dots$  
    <span style="color:red">----</span>Converge a $0$. $\lim_{n \to \infty} \frac{45 - 16n^2}{13 + 5n + 6n^3} = 0$.

(h) $b_n = \frac{3n+1}{\sqrt{4n^2+1}}, n = 1,2,3,\dots$  
    <span style="color:red">----</span>Converge a $\frac{3}{2}$. $\lim_{n \to \infty} \frac{3n+1}{\sqrt{4n^2+1}} = \frac{3}{2}$.

(i) $a_n = (-2)^{2n+1}, n = 1,2,3,\dots$  
    <span style="color:red">----</span>Diverge, ya que alterna entre valores grandes positivos y negativos.

(j) $a_n = \frac{10 - 16n^3}{1 + n^2}, n = 1,2,3,\dots$  
    <span style="color:red">----</span>Diverge, ya que $\lim_{n \to \infty} a_n = -\infty$.

(k) $a_n = \sqrt{\frac{3n^2 + n - 6}{5n^2 + 16}}, n = 1,2,3,\dots$  
    <span style="color:red">----</span>Converge a $\frac{\sqrt{3}}{\sqrt{5}}$. $\lim_{n \to \infty} a_n = \frac{\sqrt{3}}{\sqrt{5}}$.

(l) $b_n = \frac{(-1)^n}{5^n}, n = 0,1,2,\dots$  
    <span style="color:red">----</span>Converge a $0$. $\lim_{n \to \infty} b_n = 0$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de valores para n
n = np.arange(1, 50)

# Definir las secuencias que convergen
a_converge = [1/(3**n),
            (3*n - 1) / (2*n + 6),
            (2*n**5 - 3*n**2 + 23) / (7*n**5 + 13*n**4 - 12),
            (45 - 16*n**2) / (13 + 5*n + 6*n**3),
            (3*n + 1) / np.sqrt(4*n**2 + 1),
            np.sqrt((3*n**2 + n - 6) / (5*n**2 + 16)),
            (-1)**n / 5**n]

# Etiquetas para las secuencias
labels = ["a_n = 1/3^n",
        "b_n = (3n - 1) / (2n + 6)",
        "b_n = (2n^5 - 3n^2 + 23) / (7n^5 + 13n^4 - 12)",
        "c_n = (45 - 16n^2) / (13 + 5n + 6n^3)",
        "b_n = (3n + 1) / sqrt(4n^2 + 1)",
        "a_n = sqrt((3n^2 + n - 6) / (5n^2 + 16))",
        "b_n = (-1)^n / 5^n"]

# Graficar las secuencias
plt.figure(figsize=(10, 8))
for seq, label in zip(a_converge, labels):
    plt.plot(n, seq, '.', label=label)

# Añadir etiquetas y leyendas
plt.title('Secuencias que Convergen')
plt.xlabel('n (términos de la secuencia)')
plt.ylabel('Valores de a_n')
plt.legend()
plt.grid(True)
plt.show()
```

## 8.
Consider the sequence $\{a_n\}$ with $a_1 = 10$ and 

$a_{n+1} = \frac{1}{3}a_n$

for $n = 1, 2, 3, \dots$. Plot the points $(n, a_n)$ for $n = 1, 2, 3, \dots, 50$. Do you think this sequence has a limit? Can you verify this?


```python
n_values = np.arange(1, 51)
a_n = [10]

for n in range(1, len(n_values)):
    a_n.append((1/3) * a_n[-1])

plt.figure(figsize=(10, 6))
plt.plot(n_values, a_n, marker='o', linestyle='-', color='b', label='Secuencia a_n')
plt.axhline(y=0, color='r', linestyle='--', label='Límite (posible)')
plt.xlabel('n (términos de la secuencia)')
plt.ylabel('a_n (valores de la secuencia)')
plt.legend()
plt.grid(True)
plt.show()
```

La secuencia $\{a_n\}$ tiene un límite conforme $n \to \infty$. A medida que $n$ aumenta, los términos de la secuencia disminuyen rápidamente y se acercan a 0. Por lo tanto, podemos concluir que:

$\lim_{n \to \infty} a_n = 0$


## 9.
Consider the sequence $\{ a_n \}$ with $a_1 = 2$ and $a_{n+1} = 2a_n$  
for n = 1,2,3,.... Plot the points (n, a_n) for n = 1,2,3,..., 50.  
  
Can you find the limit of this sequence using the same method you used in Problem 8? Does this  
sequence have a limit?

```python

n_values = np.arange(1, 51)
a_n = [2]

for n in range(1, len(n_values)):
    a_n.append(2 * a_n[-1])

plt.figure(figsize=(10, 6))
plt.plot(n_values, a_n, 'o', color='b', label='Secuencia a_n')
plt.xlabel('n (términos de la secuencia)')
plt.ylabel('a_n (valores de la secuencia)')
plt.legend()
plt.grid(True)
plt.show()
```

La secuencia $\{a_n\}$ no tiene un límite finito. Conforme $n$ aumenta, los términos de la secuencia crecen exponencialmente hacia valores extremadamente grandes, lo que indica que la secuencia diverge. Por lo tanto, podemos concluir que:


$\lim_{n \to \infty} a_n = \infty$
